```python
import numpy as np
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt,pydotplus
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier, plot_tree
from IPython.display import Image, display
```

```python
# Load the dataset
crops_df = pd.read_csv('Crop_recommendation.csv')
crops_df
```

```python
# Using the features, define what type of crop it is
FEATURES = ['nitrogen', 'phosphorus', 'potassium', 'temperature', 'humidity','ph']
x = crops_df[FEATURES].copy()
y = crops_df['label']

dct = DecisionTreeClassifier()
dct.fit(x,y)

crops_df['prediction'] = dct.predict(crops_df[FEATURES])

crops_df
```

```python
plt.figure(figsize=(30, 30))
plot_tree(dct)
plt.show()
```
